# 3. Quais são as 10 músicas mais populares e seus respectivos artistas?

```python
top10 = df.nlargest(10, 'popularity')

for i, row in top10.iterrows():
    print(f"{row['name']} - {row['artists']} | Popularidade: {row['popularity']}")
```

Comentário: Foram selecionadas as 10 músicas com maior índice de popularidade no dataset. Em seguida, foram exibidos seus respectivos nomes e artistas. A popularidade é uma métrica do Spotify que indica o nível de sucesso e destaque de cada música na plataforma.

# 4. Qual é a duração média das músicas por gênero?

```python
duracao = (
    df.groupby('genre')['duration_ms']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,6))

plt.barh(
    duracao.index,
    duracao.values / 60000
)

plt.title('Duração Média das Músicas por Gênero')
plt.xlabel('Minutos')
plt.ylabel('Gênero')

plt.show()
```

Comentário: Foi calculada a média da duração das músicas para cada gênero musical, agrupando os dados por gênero e convertendo a duração de milissegundos para minutos. Isso permite comparar quais estilos musicais possuem músicas mais longas ou mais curtas em média.

# 5. Quais gêneros possuem músicas mais dançantes?

```python
top_dance = (
    df.groupby('genre')['danceability']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

for genre, value in top_dance.items():
    print(f"{genre}: {value:.3f}")
```

Comentário: Foi calculada a média da dançabilidade das músicas por gênero. A dançabilidade é uma métrica do Spotify que indica o quanto uma música é adequada para dança. O resultado permite identificar quais gêneros apresentam músicas mais dançantes em média.

# 6. Como a duração das músicas evoluiu ao longo dos anos?

```python
duracao_ano = (
    df.groupby('year')['duration_ms']
    .mean()
    .sort_index()
)

resultado = duracao_ano / 60000

print(resultado)
```

Comentário: As músicas foram agrupadas por ano e foi calculada a média da duração de cada período. Em seguida, esses valores foram convertidos para minutos e organizados em um gráfico de linha, permitindo analisar como a duração das músicas evoluiu ao longo do tempo.”

# 7. Quais são os artistas mais frequentes no dataset?

```python
import matplotlib.pyplot as plt

top_artistas = (
    df['artists']
    .value_counts()
    .head(10)
)

plt.figure(figsize=(12,6))

plt.barh(
    top_artistas.index,
    top_artistas.values
)

plt.title('Top 10 Artistas Mais Frequentes no Dataset')
plt.xlabel('Número de músicas')
plt.ylabel('Artista')

plt.tight_layout()
plt.show()
```

Comentário: Foi feita a contagem de quantas vezes cada artista aparece no dataset. Isso permite identificar os artistas com maior número de músicas registradas na base de dados, mostrando quais estão mais representados no conjunto analisado.

# 8. Qual é a música mais antiga e a mais recente do dataset?

```python
print("Ano mais antigo:", df['year'].min())
print("Ano mais recente:", df['year'].max())
```

```python
oldest_song = df.loc[df['year'].idxmin()]

print("🎵 Música mais antiga do dataset")
print("Ano:", oldest_song['year'])
print("Música:", oldest_song['name'])
print("Artista:", oldest_song['artists'])
```

```python
newest_song = df.loc[df['year'].idxmax()]

print("🎵 Música mais recente do dataset")
print("Ano:", newest_song['year'])
print("Música:", newest_song['name'])
print("Artista:", newest_song['artists'])
```
